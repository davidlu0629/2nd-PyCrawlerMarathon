## 1. 取出今天所有的新聞

```python
import time

t = time.localtime()
current_time = time.strftime("%Y/%m/%d %H:%M", t)
current_date = time.strftime("%Y/%m/%d", t)
print(current_time)
```

```python
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

```

```python
while(1):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    #grab the article information
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")

    d = soup.find(class_="part_list_2").find_all('h3')
    
    if d[-1].find(class_="date").text.split(' ')[0] != current_date:
        for EArticle in d:
            if EArticle.find(class_="date").text.split(' ')[0] == current_date:
                print(EArticle.find(class_="date").text, EArticle.find_all('a')[-1].text)
        break
        
    #analyze time of article to check the time
    
```

## 2. 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

pre2hr = int(current_time.split(" ")[1].split(":")[0])+22
    
browser.get("https://www.ettoday.net/news/news-list.htm")
while(1):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    #grab the article information
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")

    d = soup.find(class_="part_list_2").find_all('h3')
    
    if int(d[-1].find(class_="date").text.split(' ')[1].split(':')[0])+24 < pre2hr or (int(d[-1].find(class_="date").text.split(' ')[1].split(':')[0])+24 == pre2hr and int(d[-1].find(class_="date").text.split(' ')[1].split(':')[1])<int(current_time.split(" ")[1].split(":")[1])):
        for EArticle in d:
            if int(EArticle.find(class_="date").text.split(' ')[1].split(':')[0])+24 > pre2hr:
                print(EArticle.find(class_="date").text, EArticle.find_all('a')[-1].text)
            elif int(EArticle.find(class_="date").text.split(' ')[1].split(':')[0])+24 == pre2hr and int(EArticle.find(class_="date").text.split(' ')[1].split(':')[1])>int(current_time.split(" ")[1].split(":")[1]):
                print(EArticle.find(class_="date").text, EArticle.find_all('a')[-1].text)
        break
```
